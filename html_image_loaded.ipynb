```python
import base64
data_uri = base64.b64encode(open('image.png', 'rb').read()).decode('utf-8')
img_tag = '<img src="data:image/png;base64,{0}">'.format(data_uri)
# print(img_tag)
with open('ex.html', 'w') as f:
    f.write(img_tag)
```
