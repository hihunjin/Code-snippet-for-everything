```python
#### Create Dataframe:
import pandas as pd
 
#Create a DataFrame
d = {
    'Name':['Alisa','Bobby','jodha','jack','raghu','Cathrine',
            'Alisa','Bobby','kumar','Alisa','Alex','Cathrine'],
    'Age':[26,24,23,22,23,24,26,24,22,23,24,24]
}
df = pd.DataFrame(d,columns=['Name','Age'])
df
```

# Check a column has unique values

```python
if df['Name'].duplicated().sum() > 0:
    print('duplicated')
else:
    print('unique values')
```

# Remove duplicated rows

```python
df.drop_duplicates(
    ignore_index=True # default False
)
```
