```python
import numpy as np
from PIL import Image

imarray = np.random.rand(100,100,3) * 255
im = Image.fromarray(imarray.astype('uint8')).convert('RGBA')
```

```python
im
```

```python
im.size
```

```python
out = im.crop((-3, -3, 10, 10))
out2 = im.crop((0, 0, 13, 13))
```
