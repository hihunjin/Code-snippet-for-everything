```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
out = np.random.rand(100,50)#1000
out3 = out * 100
print(out3.min(), out3.max())
```

```python
plt.figure()
plt.imshow(out3, cmap='jet')
ticks = np.linspace(out3.min(), out3.max(), 10, endpoint=True)
plt.colorbar(
    ticks=ticks, 
    shrink=0.7,
)
plt.axis('off')
plt.savefig('hi.png')
```
