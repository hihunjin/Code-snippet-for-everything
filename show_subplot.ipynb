```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
%%capture
!wget https://thispersondoesnotexist.com -O 'image.png'
```

```python
fig, axs = plt.subplots(2, 3, figsize=(10,10))
fig.suptitle('Vertically stacked subplots')
axs[0,0].imshow(Image.open('image.png'))
axs[0,1].imshow(Image.open('image.png'))
axs[0,1].axis('off')
axs[0,2].axis('off')
axs[1,0].imshow(Image.open('image.png'))
axs[1,1].imshow(Image.open('image.png'))
axs[1,1].axis('off')
axs[1,2].imshow(Image.open('image.png'))
```

```python
fig, axs = plt.subplots(2, 3, gridspec_kw = {'wspace':0, 'hspace':0})
fig.suptitle('Vertically stacked subplots')
axs[0,0].imshow(Image.open('image.png'))
axs[0,0].axis('off')
axs[0,1].imshow(Image.open('image.png'))
axs[0,1].axis('off')
axs[0,2].axis('off')
axs[1,0].imshow(Image.open('image.png'))
axs[1,0].axis('off')
axs[1,1].imshow(Image.open('image.png'))
axs[1,1].axis('off')
axs[1,2].imshow(Image.open('image.png'))
axs[1,2].axis('off')
plt.tight_layout()
```
