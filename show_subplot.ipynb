```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,10))
fig.suptitle('Vertically stacked subplots')
axs[0,0].imshow(Image.open('image.png'))
axs[0,1].imshow(Image.open('image.png'))
axs[1,0].imshow(Image.open('image.png'))
axs[1,1].imshow(Image.open('image.png'))
```
